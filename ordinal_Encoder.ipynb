```python
import pandas as pd
```

```python
d = {'sales': [100000, 222000, 1000000, 522000, 111111, 222222, 1111111, 20000, 75000, 90000, 1000000, 10000],
     'city': ['Tampa', 'Tampa', 'Orlando', 'Jacksonville', 'Miami', 'Jacksonville', 'Miami', 'Miami', 'Orlando',
              'Orlando', 'Orlando', 'Orlando'],
     'size': ['Small', 'Medium', 'Large', 'Large', 'Small', 'Medium', 'Large', 'Small', 'Medium', 'Medium', 'Medium',
              'Small', ]}
```

```python
df = pd.DataFrame(data=d)
df.head()
```

```python
df['size'].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
sizes = ['Small', 'Medium', 'Large']
```

```python
enc = OrdinalEncoder(categories =[sizes])
```

```python
df['size'] =enc.fit_transform(df[['size']])
```
