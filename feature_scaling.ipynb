```python
import pandas as pd
df = pd.read_csv('dataset/Exam_Score_prediction.csv')
```

```python
newdf = df.drop(columns =['gender','course','internet_access','exam_difficulty','facility_rating','study_method','sleep_quality'])
```

```python
newdf
```

```python
X1 = newdf.iloc[:,:6]
```

```python
X2 = newdf.iloc[:,:6]
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler_std = StandardScaler()
scaler_min_max = MinMaxScaler(feature_range = (0,1))
```

```python
X1 = scaler_std.fit_transform(X1)
X2 = scaler_min_max.fit_transform(X2)
```

```python
X1 = pd.DataFrame(X1,columns = ['student_id','age','study_hrs','class_attendance','sleep_hrs','exam_score'])
```

```python
X1.describe().round(3)
```
